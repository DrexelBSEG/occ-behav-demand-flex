```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data from csv file
df = pd.read_csv('PPD-PMV.csv')
```

```python
# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 5))

# Define bar width and position
bar_width = 0.35
index = df["PMV"].values
rp884_bars = ax.bar(index - bar_width/2, df["RP884"]*100, bar_width, label='RP-884', color='#1f77b4')
case_study_bars = ax.bar(index + bar_width/2, df["Case Study"]*100, bar_width, label='Sampled Occupants', color='#ff7f0e')

# Set up the plot details
ax.set_xlabel("PMV", fontsize=16)
ax.set_ylabel("Predicted Percentage of Dissatisfied (PPD)", fontsize=16)
ax.set_xticks(index)
ax.legend(fontsize=14)
ax.grid(axis='y')
ax.set_ylim([0, 100])

# Increase tick font
ax.tick_params(axis='both', which='major', labelsize=14)

plt.tight_layout()
plt.show()
```
